# **Importing dependencies**

```python
from numpy import loadtxt #handle/load data
import tensorflow
from keras.models import Sequential #empty working area
from keras.layers import Dense #dense layer
```

# **Data collection**

```python
dataset = loadtxt('/content/diabetes.csv', delimiter = ',')
```

# **Data splitting**

```python
X = dataset[:, 0:8]
Y = dataset[:,8]
```

```python
X
```

```python
Y
```

# **Training the model**

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation = 'relu')) #input layer
model.add(Dense(8, activation = 'relu')) #output layer
model.add(Dense(1, activation = 'sigmoid')) #output layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics = ['accuracy'])
model.fit(X, Y, epochs=5, batch_size=10)
```

# **Evaluating the model**

```python
accuracy = model.evaluate(X, Y)
print('Accuracy: ', accuracy*100)
```

# **Saving the data**

```python
#saving the data using JSON
model_json = model.to_json()
with open("model.json", "w") as json_file:
  json_file.write(model_json)

model.save_weights('model.h5')
print("Saved model to disk")
```
